1. What is the role of try and exception block?

it is a mechanisem in programming to handle runtime errors also known as exceptions. In python the try block contains the code that might cause an error and except block contains the code to run if an error occours.

2. What is the syntax for a basic try-except block?

```python
#a very basic try and except block

try:
  print(x)
except:
  print("Some issue with x")
```

exception type?

bare except blocks are used to stop program from terminating suddenly they can catch errors but can't tell which type of error.
when we specify a exeption type it helps us deal with only that particular exception/error in a more controlled andmeaning ful way


5. Can you have nested try-except blocks in Python? If yes, then give an example.

```python
#yes we can have nested try-except blocks in python

def divide_numbers(a, b):
    try:
        res = a / b
    except ZeroDivisionError:
        print("Error: Division by zero")
    else:
        try:
            print("The result is:", res)
        except NameError:
            print("Error Occoured: ")
            
divide_numbers(5,0), divide_numbers(5,100)
```

6. Can we use multiple exception blocks, if yes then give an example.

```python
#example of multiple except blocks

try:
    n = int(input("Please enter the numerator: "))
    d = int(input("Please enter the denominator: "))

    result = n / d
    print("Result:", result)

except ValueError:
    print("Please enter valid integers for the numerator and denominator.")

except ZeroDivisionError:
    print("Division by zero is not allowed.")
    
```

7. Write the reason due to which following errors are raised:
- a. EOFError
- b. FloatingPointError
- c. IndexError
- d. MemoryError
- e. OverflowError
- f. TabError
- g. ValueError

a. EOFError: This error is raised when a built-in function like input() or a file reading operation reaches the end of the file, but further data is expected. It indicates that there is an attempt to read beyond the end of the file.

b. FloatingPointError: This error is raised when a floating-point operation fails to produce a valid result. It can occur when performing calculations involving floating-point numbers, such as division by zero or attempting to calculate the square root of a negative number.

c. IndexError: This error is raised when an index is out of range in a sequence (e.g., list, string, tuple). It occurs when you try to access an element using an invalid index that is either too large or too small.

d. MemoryError: This error is raised when the computer's memory allocation fails. It typically occurs when a program tries to allocate more memory than the system can provide. This error indicates that there is insufficient memory available to perform the requested operation.

e. OverflowError: This error is raised when the result of an arithmetic operation exceeds the maximum representable value for a numeric type. It occurs when performing calculations that produce a value outside the range of values that can be stored in the given data type.

f. TabError: This error is raised when there is an issue with the indentation in Python code. It occurs when inconsistent use of tabs and spaces is encountered, typically within the same block of code.

g. ValueError: This error is raised when a function receives an argument of the correct type but an inappropriate value. It occurs when an operation or function is provided with an argument that is of the correct data type but is not a valid value for that operation or function.

8. Write code for the following given scenario and add try-exception block to it.
- a. Program to divide two numbers

```python
try:
    num1 = float(input("Enter the numerator: "))
    num2 = float(input("Enter the denominator: "))
    result = num1 / num2
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```

```python
try:
    string_num = input("Enter a number: ")
    num = int(string_num)
    print("Number:", num)
except ValueError:
    print("Error: Invalid input. Please enter a valid integer.")
```

```python
try:
    my_list = [1, 2, 3, 4, 5]
    index = int(input("Enter the index: "))
    element = my_list[index]
    print("Element at index", index, ":", element)
except IndexError:
    print("Error: Index is out of range.")
```
