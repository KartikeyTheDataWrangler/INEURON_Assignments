2. Can a try-except block be nested inside another try-except block? Explain with an
example.

```python

def divide_numbers(a, b):
    try:
        res = a / b
    except ZeroDivisionError:
        print("Error: Division by zero")
    else:
        try:
            print("The result is:", res)
        except NameError:
            print("Error Occoured: ")
            
divide_numbers(5,0), divide_numbers(5,100)
```

```python
import logging

logging.basicConfig(level=logging.WARNING)

def MyBalance(amount):
    if amount < 40000:
        logging.warning('Sorry you have Low balance: %s', amount)

MyBalance(10000)
```

4. ERROR : Errors that will definately cause the some part of program break hence will nedd to be fixed 

```python
import logging

logging.basicConfig(level=logging.ERROR)

def LetUsDivide(n, d):
    try:
        result = n / d
    except ZeroDivisionError:
        logging.error('You are trying to divide by zero, which is not allowed')
    else:
        return result

LetUsDivide(4, 0)
```

5. Critical Errors: these are the errors which will cause the whole software to stop functioning


```python
import logging

logging.basicConfig(level=logging.CRITICAL)

def LetUsCheckSystem(sys):
    if sys != 'OK':
        logging.critical('System failure: %s', sys)

LetUsCheckSystem('You need to handle the issue now')
```

7. What are log formatters in Python logging, and how can you customise the log
message format using formatters?

In Python logging, log formatters are used to specify the format of log messages. They determine how log records are transformed into a string representation that can be displayed or saved. Log formatters provide flexibility in customizing the content and structure of log messages, allowing you to include relevant information such as timestamps, log levels, module names, and more.

```python
import logging

formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')

logger = logging.getLogger('my_logger')
handler = logging.StreamHandler()
handler.setFormatter(formatter)
logger.addHandler(handler)

logger.info('This is some information')
```

8. How can you set up logging to capture log messages from multiple modules or
classes in a Python application?

example of code for setting up logger to get logging multiple classes 

```python
import logging

logging.basicConfig(level=logging.INFO)

class Car:
    def __init__(self):
        self.logger = logging.getLogger('Car')

    def perform_action(self):
        self.logger.info('Some action performed on car')


class Person:
    def __init__(self):
        self.logger = logging.getLogger('Person')

    def perform_action(self):
        self.logger.info('Action performed in person')
```

```python
class1 = Car()
class2 = Person()
class1.perform_action()
class2.perform_action()
```

10. Write a Python program that logs a message to a file named "app.log" with the
following requirements:
- The log message should be "Hello, World!"
- The log level should be set to "INFO."
- The log file should append new log entries without overwriting previous ones.

```python
import logging

# Configure logging to write to a file
logging.basicConfig(filename='app.log', level=logging.INFO, filemode='a', format='%(asctime)s - %(levelname)s - %(message)s')

# Log the message
logging.info("Hello, World!")
```

11. Create a Python program that logs an error message to the console and a file named
"errors.log" if an exception occurs during the program's execution. The error
message should include the exception type and a timestamp.

```python
import logging
import traceback
import sys

#logger object
logger = logging.getLogger('error_logger')
logger.setLevel(logging.ERROR)

# handler for showing errors in console
console_handler = logging.StreamHandler()
console_handler.setLevel(logging.ERROR)

# file handler for writing to a log file
file_handler = logging.FileHandler('errors.log')
file_handler.setLevel(logging.ERROR)

#creating a formatter 
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
console_handler.setFormatter(formatter)
file_handler.setFormatter(formatter)


logger.addHandler(console_handler)
logger.addHandler(file_handler)

try:
    #some code
    raise ValueError("An error occurred!")

except Exception as e:
    # Log the error message
    exc_type, exc_value, exc_traceback = sys.exc_info()
    error_msg = f"Exception: {exc_type.__name__} - {exc_value}\n{traceback.format_exc()}"
    logger.error(error_msg)


```
