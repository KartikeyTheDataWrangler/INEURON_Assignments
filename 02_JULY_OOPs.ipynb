4. What is the significance of the "protected" access modifier in inheritance? How does
it differ from "private" and "public" modifiers?

Public modifiers is the default in which new classes are made in python. Public class is accessible from outside the class as well as from the child classes

```python
class MC:
    def __init__(self):
        self.pv = 30  
    def pm(self):
        print("This is a public method.")

obj = MC()
print(obj.pv)  
obj.pm()  
```

Protected variables are accessable from their own class as well as a derived classes but not from any other class (class not inheriting from the base) 

```python
#making a protected class
class base:
    def __init__(self):
        pass
        
        
    def _fruits(self):                       #making a protected class using single underscore
        print("some method in base class")
        
class derived(base):
    def printfruits(self):
          self._fruits()                 #derived class accessing the protected method
    
der1 = derived()
der1.printfruits()
```

private modifiers are only accessible within the same class where they are declared.
They cannot be accessed directly from derived classes or other parts of the program.

```python
class MyClass:
    def __init__(self):
        self.__private_var = 10  

    def __private_method(self):
        print("This is a private method.")  #private method defined using double undersocres

obj = MyClass()
print(obj.__private_var)  # Error: Private variable is not accessible
obj.__private_method()  # Error: Private method is not accessible
```

5. What is the purpose of the "super" keyword in inheritance? Provide an example.

super() keyword is used in a derived class to inherit the method or coonstructor from parent class

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        print(f"The {self.brand} vehicle is starting.")


class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def start(self):
        super().start()  
        print(f"The {self.brand} {self.model} is ready to go.")


car = Car("Toyota", "Innova")
car.start()
```

6. Create a base class called "Vehicle" with attributes like "make", "model", and "year".
Then, create a derived class called "Car" that inherits from "Vehicle" and adds an
attribute called "fuel_type". Implement appropriate methods in both classes.

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

class Car(Vehicle):
    def __init__(self,model,year,make,fuel_type):  #constructor of vehicle
        super().__init__(make,model,year)            #inheriting from vehicle
        self.fuel_type = fuel_type
        
    def display_car(self):
        print(f"the cars model is {self.model} \nand it was made in {self.year} \nthe make is {self.make} \nuses fuel type {self.fuel_type}")
        
```

```python
car1 = Car('VERNA','2017','proplus','diesel')
car1.display_car()
```

7. Create a base class called "Employee" with attributes like "name" and "salary."
Derive two classes, "Manager" and "Developer," from "Employee." Add an additional
attribute called "department" for the "Manager" class and "programming_language"
for the "Developer" class.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def basic_details(self):
        print(f"The Employees name is {self.name} and salary is {self.salary}")

class Manager(Employee):
    def __init__(self,department,name):
        super().__init__(name)
        self.department = department
    
    def dis_department(self):
        print(f"Employee {self.name} works at department {self.department}")
    
class Developer(Employee):
    def __init__(self,name,salary,programming_language):
        super().__init__(name,salary)
        self.programming_language = programming_language
        
    def language(self):
        print(f"Employee {self.name} uses programming language {self.programming_language}")          
```

```python
emp1 = Developer("Beleth","10000","python")
emp1.basic_details()
emp1.language()
```

8. Design a base class called "Shape" with attributes like "colour" and "border_width."
Create derived classes, "Rectangle" and "Circle," that inherit from "Shape" and add
specific attributes like "length" and "width" for the "Rectangle" class and "radius" for
the "Circle" class.


```python
class Shape:
    def __init__(self,colour,border_width):
        self.colour = colour
        self.border_width = border_width
        
class Rectangle(Shape):
    def __init__(self,length,width,colour):
        super().__init__(colour,border_width=None)
        self.length = length
        self.width = width

    def rect_det(self):
        print(f"the given rectangle has colour {self.colour} and length {self.length}")
        

class Circle(Shape):
    def __init__(self,colour,radius):
        super().__init__(colour,border_width=None)
        self.radius = radius
        
    def cir_det(self):
        print(f"the given circle has colour {self.colour} and radius {self.radius}")
```

```python
rec1 = Rectangle(150,200,'Orange')
rec1.rect_det()
cir1 = Circle('Red',50)
cir1.cir_det()
```

9. Create a base class called "Device" with attributes like "brand" and "model." Derive
two classes, "Phone" and "Tablet," from "Device." Add specific attributes like
"screen_size" for the "Phone" class and "battery_capacity" for the "Tablet" class.


```python
class Device:
    def __init__(self,model,brand=None):
        self.brand = brand
        self.model = model
        
class Phone(Device):
    def __init__(self,brand,model,screen_size):
        super().__init__(model,brand)
        self.screen_size = screen_size
    
    def some_detail(self):
        print(f"the screen size of the {self.brand} : {self.model} is {self.screen_size}")
        
        
class Tablet(Device):
    def __init__(self,model,battery_capacity):
        super().__init__(model)
        self.battery_capacity = battery_capacity
        
    def show_battery(self):
        print(f"the battery capacity for {self.model} is {self.battery_capacity}")
    
```

```python
dev1 = Tablet('Redmi note 3','4000 Mah')
dev1.show_battery()
```

10. Create a base class called "BankAccount" with attributes like "account_number" and
"balance." Derive two classes, "SavingsAccount" and "CheckingAccount," from
"BankAccount." Add specific methods like "calculate_interest" for the
"SavingsAccount" class and "deduct_fees" for the "CheckingAccount" class.

```python
class BankAccount:
    def __init__ (self,account_number,balance):
        self.account_number = account_number
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        
    def withdrawn(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else: 
            print("not enough balance in account")
            
    def display_balance(self):
        print(f"Account Number: {self.account_number}")
        print(f"Balance: {self.balance}")
        
class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance):
        super().__init__(account_number, balance)

    def calculate_interest(self, interest_rate):
        interest = self.balance * (interest_rate / 100)
        self.deposit(interest)                  #method from base class is called using self

class CheckingAccount(BankAccount):
    
    def __init__(self, account_number, balance):
        super().__init__(account_number, balance)
    
    def deduct_fees(self, fee_amount):
        self.withdraw(fee_amount)

            
        
```

```python
savings_acc = SavingsAccount("Beleth", 1000)
savings_acc.calculate_interest(2.5)
savings_acc.display_balance()
```
